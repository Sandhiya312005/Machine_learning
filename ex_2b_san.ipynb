Importing the libraries

```python
import pandas as pd
import numpy as np
```

Importing dataset

```python
d=pd.read_csv("/content/drive/MyDrive/datasets/houseprice - houseprice (1).csv")
d.head()
```

Checking null values

```python
d.isna().sum()
```

Handling null values

```python
d=d.replace('?',np.nan)
```

```python
d=d.fillna(d.mean())
```

```python
d=d.dropna()
```

Checking correlation

```python
import seaborn as sns
sns.heatmap(d.corr())
```

```python
d.corr()
```

Splitting into target and Features

```python
x=d[['sqft_living','grade']]
y=d[['price']]
```

Training linear regression model

```python
from sklearn.linear_model import LinearRegression
lr1=LinearRegression()
lr1.fit(x.iloc[:,0:1],y)
```

```python
lr2=LinearRegression()
lr2.fit(x.iloc[:,1:2],y)
```

Plotting linear regression

```python
import matplotlib.pyplot as plt
plt.scatter(x.iloc[:,0:1],y)
plt.plot(np.array(x.iloc[:,0:1]),lr1.predict(x.iloc[:,0:1]),color='Red')
```

```python
plt.scatter(x.iloc[:,1:2],y)
plt.plot(np.array(x.iloc[:,1:2]),lr2.predict(x.iloc[:,1:2]),color = "Red")
```

Calculating accuracy

```python
from sklearn.metrics import r2_score,mean_squared_error
print("Mean Sqaure Error Feature 1 : ",mean_squared_error(y,lr1.predict(x.iloc[:,0:1])))
print("Linear Regression Feature 1 R2 Score : ",r2_score(y,lr1.predict(x.iloc[:,0:1])))
print(mean_squared_error(y,lr2.predict(x.iloc[:,1:2])))
print("Mean Sqaure Error Feature 2 : ",r2_score(y,lr2.predict(x.iloc[:,1:2])))
```

Increasing degree to 3 and training the model

```python
from sklearn.preprocessing import PolynomialFeatures
p=PolynomialFeatures(degree=3)
x_poly=p.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_poly,y)
```

Calculating accuracy for Polynomial Regression

```python
from sklearn.metrics import r2_score,mean_squared_error
print(mean_squared_error(y,lr.predict(x_poly)))
r2_score(y,lr.predict(x_poly))

```

Plotting Polynomial Regression

```python
import matplotlib.pyplot as plt
x1_range=np.linspace(min(x.iloc[:,0]),max(x.iloc[:,0]),10)
x2_range=np.linspace(min(x.iloc[:,1]),max(x.iloc[:,1]),10)
x1_grid,x2_grid=np.meshgrid(x1_range,x2_range)
x_grid_poly=p.transform(np.c_[x1_grid.ravel(),x2_grid.ravel()])
y_grid=lr.predict(x_grid_poly).reshape(x1_grid.shape)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x.iloc[:,0],x.iloc[:,1],y)
ax.plot_surface(x1_grid,x2_grid,y_grid)
plt.title("717822I249 - Sandhiya P")

```
